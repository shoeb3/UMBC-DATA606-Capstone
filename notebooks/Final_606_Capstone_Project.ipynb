## Data Cleaning

We filter to H-1B petitions only, create the target variable, drop columns with too much missing data, convert dates and wages to numeric formats, and handle outliers.

```python
# number of visa classes
print("Unique VISA_CLASS values:")
print(df['VISA_CLASS'].value_counts())

# only H1b class kept
df = df[df['VISA_CLASS'] == 'H-1B'].copy()

# Show how many rows are left after filtering
print("\nAfter filtering for H-1B:")
print(len(df), "rows")
```

```python
print("Original CASE_STATUS distribution:")
print(df['CASE_STATUS'].value_counts())

df['TARGET'] = (df['CASE_STATUS'] == 'Certified').astype(int)

print("\nBinary TARGET created:")
target_counts = df['TARGET'].value_counts()
target_counts.index = ['Not Certified (0)', 'Certified (1)']

print(target_counts)
```

```python
# Identifying columns above the 50% threshold
mnullcol = report[report['Percentage'] > 50].index.tolist()

# Defining specific columns to remove for model clarity
col_id = ['CASE_NUMBER', 'VISA_CLASS', 'PREPARER_LAST_NAME', 'PREPARER_FIRST_NAME', 'PREPARER_MIDDLE_INITIAL']

# Combining and dropping columns
cld = set(mnullcol + col_id)
df.drop(columns=[c for c in cld if c in df.columns], inplace=True)

print(f"Dropped {len(cld)} columns.")
print(f"New DataFrame shape: {df.shape}")
```

```python
# Convert Date Columns
date_cols = ['RECEIVED_DATE', 'DECISION_DATE', 'BEGIN_DATE', 'END_DATE']

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Convert Wage Columns
wage_cols = ['WAGE_RATE_OF_PAY_FROM', 'WAGE_RATE_OF_PAY_TO', 'PREVAILING_WAGE']

for col in wage_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('$', '', regex=False)
        df[col] = df[col].astype(str).str.replace(',', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')


all_cols = date_cols + wage_cols
ac_cols = [c for c in all_cols if c in df.columns]
print("Dates and Wages are now numbers.")
print(df[ac_cols].dtypes)
```

```python
unit_map = {
    'Year'      : 1,
    'Month'     : 12,
    'Bi-Weekly' : 26,
    'Week'      : 52,
    'Hour'      : 2080,
}

if 'WAGE_UNIT_OF_PAY' in df.columns:
    df['WAGE_MULTIPLIER'] = df['WAGE_UNIT_OF_PAY'].map(unit_map).fillna(1)
    df['ANNUAL_WAGE'] = df['WAGE_RATE_OF_PAY_FROM'] * df['WAGE_MULTIPLIER']
    df['ANNUAL_PREVAILING_WAGE'] = df['PREVAILING_WAGE']
else:
    df['ANNUAL_WAGE'] = df['WAGE_RATE_OF_PAY_FROM']
    df['ANNUAL_PREVAILING_WAGE'] = df['PREVAILING_WAGE']

print(" Annual wages computed.")
print(df[['ANNUAL_WAGE', 'ANNUAL_PREVAILING_WAGE']].describe())
```

```python
# Wage ratio: offered vs prevailing
df['WAGE_RATIO'] = df['ANNUAL_WAGE'] / df['ANNUAL_PREVAILING_WAGE']

# Processing time in days
df['PROCESSING_DAYS'] = (df['DECISION_DATE'] - df['RECEIVED_DATE']).dt.days

# Full-time flag (1 = Yes, 0 = No)
df['IS_FULL_TIME'] = (df['FULL_TIME_POSITION'] == 'Y').astype(int)

# Wage above prevailing flag
df['WAGE_ABOVE_PREVAILING'] = (df['ANNUAL_WAGE'] > df['ANNUAL_PREVAILING_WAGE']).astype(int)

# Quick check of new features
print("New features created")
print(df[['WAGE_RATIO', 'PROCESSING_DAYS', 'IS_FULL_TIME', 'WAGE_ABOVE_PREVAILING']].describe())
```

```python
# Get 1% and 99% wage limits
low_limit = df['ANNUAL_WAGE'].quantile(0.01)
high_limit = df['ANNUAL_WAGE'].quantile(0.99)

print(f"Wages below ${low_limit:,.0f} and above ${high_limit:,.0f} are outliers.")

# Cap extreme wage values
df['ANNUAL_WAGE_CAPPED'] = df['ANNUAL_WAGE'].clip(lower=low_limit, upper=high_limit)

# Dataset size
print(f"\nFinal dataset: {len(df):,} rows, {df.shape[1]} columns")

# Count outcomes
tcases = len(df)
certified = df['TARGET'].sum()
denied = tcases - certified

# Calculate percentages
perc_certified = (certified / tcases) * 100
perc_denied = (denied / tcases) * 100

print(f"Certified: {certified:,} ({perc_certified:.1f}%)")
print(f"Denied: {denied:,} ({perc_denied:.1f}%)")
```

## Feature Engineering

Numeric features alone have weak predictive power because the DOL certifies most petitions regardless of wage. The strongest signals come from employer history and occupation patterns. We create new features in four categories: corrected wage features that fix an annualization bug, employer and occupation certification rates from historical filings, attorney-level features since some attorneys have better track records, and new signals like wage vs occupation median and worksite-employer state mismatch.

### Fixing the Prevailing Wage Bug

The original EDA had a bug where PREVAILING_WAGE was not annualized before computing the wage ratio. Many prevailing wages are stored as hourly values. Dividing an annual offered wage of 100,000 dollars by an hourly rate of 50 gives a ratio of 2,000 instead of the correct 1.05. We fix this by converting prevailing wages to annual using PW_UNIT_OF_PAY, then add log-transformed wages and a wage difference feature.

```python
unit_map = {'Year':1, 'Month':12, 'Bi-Weekly':26, 'Week':52, 'Hour':2080}

df['PW_MULTIPLIER']          = df['PW_UNIT_OF_PAY'].map(unit_map).fillna(1)
df['ANNUAL_PREVAILING_WAGE'] = df['PREVAILING_WAGE'] * df['PW_MULTIPLIER']

low_pw, high_pw = df['ANNUAL_PREVAILING_WAGE'].quantile([0.01, 0.99])
df['ANNUAL_PREVAILING_WAGE_CAP'] = df['ANNUAL_PREVAILING_WAGE'].clip(low_pw, high_pw)

df['WAGE_DIFF'] = df['ANNUAL_WAGE_CAPPED'] - df['ANNUAL_PREVAILING_WAGE_CAP']
df['LOG_WAGE']  = np.log1p(df['ANNUAL_WAGE_CAPPED'])
df['LOG_PW']    = np.log1p(df['ANNUAL_PREVAILING_WAGE_CAP'])

print("Wage features engineered.")
print(f"WAGE_DIFF corr with TARGET : {df['WAGE_DIFF'].corr(df['TARGET']):.4f}")
print(f"LOG_WAGE  corr with TARGET : {df['LOG_WAGE'].corr(df['TARGET']):.4f}")
print(f"LOG_PW    corr with TARGET : {df['LOG_PW'].corr(df['TARGET']):.4f}")
```

### Employer, Occupation, and State Certification Rates

For each employer, occupation, and state we compute the historical certification rate — what fraction of their past petitions were certified. Employers with a history of denials tend to receive future denials, and this signal has a correlation of 0.57 with the target, making it the strongest feature in the dataset.

These rates are computed on the full dataset here for display purposes only. In the ML pipeline they are recomputed using training data only to prevent data leakage. The full-dataset versions are dropped after the correlation matrix is shown.

```python
emp_stats = df.groupby('EMPLOYER_NAME')['TARGET'].agg(['mean', 'count']).reset_index()
emp_stats.columns = ['EMPLOYER_NAME', 'EMP_CERT_RATE', 'EMP_PETITION_COUNT']

soc_stats = df.groupby('SOC_TITLE')['TARGET'].agg(['mean', 'count']).reset_index()
soc_stats.columns = ['SOC_TITLE', 'SOC_CERT_RATE', 'SOC_PETITION_COUNT']

state_stats = df.groupby('EMPLOYER_STATE')['TARGET'].agg(['mean']).reset_index()
state_stats.columns = ['EMPLOYER_STATE', 'STATE_CERT_RATE']

df = df.merge(emp_stats,   on='EMPLOYER_NAME',  how='left')
df = df.merge(soc_stats,   on='SOC_TITLE',      how='left')
df = df.merge(state_stats, on='EMPLOYER_STATE',  how='left')

emp_median = df.groupby('EMPLOYER_NAME')['ANNUAL_WAGE_CAPPED'].median()
df['EMP_MEDIAN_WAGE']  = df['EMPLOYER_NAME'].map(emp_median)
df['WAGE_VS_EMP_NORM'] = df['ANNUAL_WAGE_CAPPED'] - df['EMP_MEDIAN_WAGE']

df['IS_LARGE_EMPLOYER'] = (df['EMP_PETITION_COUNT'] > 100).astype(int)

print('Employer, occupation, and state features created.')
print(f"EMP_CERT_RATE    corr: {df['EMP_CERT_RATE'].corr(df['TARGET']):.4f}")
print(f"SOC_CERT_RATE    corr: {df['SOC_CERT_RATE'].corr(df['TARGET']):.4f}")
print(f"STATE_CERT_RATE  corr: {df['STATE_CERT_RATE'].corr(df['TARGET']):.4f}")
print(f"IS_LARGE_EMPLOYER corr: {df['IS_LARGE_EMPLOYER'].corr(df['TARGET']):.4f}")
```

### Updated Correlation Matrix

After adding the rate features the correlation picture changes significantly. EMP_CERT_RATE at 0.57 is now the dominant signal, followed by SOC_CERT_RATE at 0.22 and ATTY_CERT_RATE at 0.25. These features drive the machine learning models.

```python
new_feat_cols = ['TARGET', 'ANNUAL_WAGE_CAPPED', 'WAGE_DIFF', 'LOG_WAGE', 'LOG_PW',
                 'EMP_CERT_RATE', 'SOC_CERT_RATE', 'STATE_CERT_RATE',
                 'IS_LARGE_EMPLOYER', 'EMP_PETITION_COUNT']

corr_new = df[new_feat_cols].corr()

plt.figure(figsize=(11, 8))
sns.heatmap(corr_new, annot=True, fmt='.2f', cmap='RdBu', center=0, linewidths=0.5)
plt.title('Updated Correlation Matrix — With New Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('Correlations with TARGET (sorted):')
print(corr_new['TARGET'].drop('TARGET').sort_values(ascending=False).round(4))

df.drop(columns=['EMP_CERT_RATE', 'SOC_CERT_RATE', 'STATE_CERT_RATE'], inplace=True)
print('\nLeaky rate columns dropped from df. Will be recomputed train-only in ML phase.')
```

## Machine Learning

The goal is to predict whether an H-1B petition will be certified or denied using only information available at the time of filing. We take a stratified 150,000 row sample, split 80/20 for training and testing, recompute certification rates on training data only to avoid leakage, apply smoothed target encoding to categorical columns, and train three models: Logistic Regression, Random Forest, and XGBoost.

Since 92.6% of petitions are certified, accuracy alone is misleading. We focus on denial recall and AUC-ROC.

### Libraries

### Data Split and Leakage-Free Features

We take a stratified 150,000 row sample preserving the class ratio. EMP_CERT_RATE, SOC_CERT_RATE, and STATE_CERT_RATE are computed after the split using only training labels, bringing the employer rate correlation from 0.57 down to 0.37, which is the honest value without leakage.

New features added here include PW_LEVEL for the prevailing wage tier, HAS_ATTORNEY and ATTY_CERT_RATE for attorney history, STATE_MISMATCH for cases where the worksite is in a different state than the employer, and WAGE_VS_SOC_MEDIAN for how the wage compares to the occupation median. PROCESSING_DAYS is excluded because it is only known after the decision is made.

```python
# Base features — rate columns excluded here, recomputed after split
feat_base = [
    'ANNUAL_WAGE_CAPPED',
    'ANNUAL_PREVAILING_WAGE_CAP',
    'WAGE_ABOVE_PREVAILING',
    'IS_FULL_TIME',
    'QUARTER',
    'WAGE_DIFF',
    'LOG_WAGE',
    'LOG_PW',
    'IS_LARGE_EMPLOYER',
    'EMP_PETITION_COUNT',
    'EMPLOYER_NAME',
    'SOC_TITLE',
    'EMPLOYER_STATE',
]

EXTRA_COLS  = ['AGENT_ATTORNEY_LAST_NAME', 'PW_WAGE_LEVEL', 'WORKSITE_STATE']
extra_avail = [c for c in EXTRA_COLS if c in df.columns]

df_model = df[feat_base + extra_avail + ['TARGET']].dropna(subset=feat_base + ['TARGET'])

df_sample = df_model.groupby('TARGET', group_keys=False).apply(
    lambda x: x.sample(min(len(x), int(150000 * len(x) / len(df_model))), random_state=42))

X_train, X_test, y_train, y_test = train_test_split(
    df_sample.drop(columns=['TARGET']),
    df_sample['TARGET'],
    test_size=0.2, random_state=42, stratify=df_sample['TARGET'])

print(f'Sample : {len(df_sample):,} rows')
print(f'Train  : {len(X_train):,} rows , Certified: {y_train.mean()*100:.1f}%')
print(f'Test   : {len(X_test):,}  rows , Certified: {y_test.mean()*100:.1f}%')

print('\nComputing rate features on training set only')
global_mean = float(y_train.mean())

def smoothed_rate(train_x, train_y, test_x, group_col, smoothing=50):
    tmp = train_x[[group_col]].copy()
    tmp['_y'] = train_y.loc[train_x.index].values  # index-aligned
    agg = tmp.groupby(group_col)['_y'].agg(['mean', 'count'])
    rate_map = ((agg['count'] * agg['mean'] + smoothing * global_mean) /
                (agg['count'] + smoothing)).to_dict()
    return (train_x[group_col].map(rate_map).fillna(global_mean),
            test_x[group_col].map(rate_map).fillna(global_mean))

X_train, X_test = X_train.copy(), X_test.copy()

X_train['EMP_CERT_RATE'],   X_test['EMP_CERT_RATE']   = smoothed_rate(X_train, y_train, X_test, 'EMPLOYER_NAME')
X_train['SOC_CERT_RATE'],   X_test['SOC_CERT_RATE']   = smoothed_rate(X_train, y_train, X_test, 'SOC_TITLE')
X_train['STATE_CERT_RATE'], X_test['STATE_CERT_RATE'] = smoothed_rate(X_train, y_train, X_test, 'EMPLOYER_STATE')

print(f"EMP_CERT_RATE   corr: {X_train['EMP_CERT_RATE'].corr(y_train):.4f}")
print(f"SOC_CERT_RATE   corr: {X_train['SOC_CERT_RATE'].corr(y_train):.4f}")
print(f"STATE_CERT_RATE corr: {X_train['STATE_CERT_RATE'].corr(y_train):.4f}")

# New features
print('\nBuilding additional features')

# PW_WAGE_LEVEL Level I = entry-level, highest denial rate
if 'PW_WAGE_LEVEL' in X_train.columns:
    pw_map = {'I': 1, 'II': 2, 'III': 3, 'IV': 4}
    X_train['PW_LEVEL'] = X_train['PW_WAGE_LEVEL'].map(pw_map).fillna(2)
    X_test['PW_LEVEL']  = X_test['PW_WAGE_LEVEL'].map(pw_map).fillna(2)
    X_train.drop(columns=['PW_WAGE_LEVEL'], inplace=True)
    X_test.drop(columns=['PW_WAGE_LEVEL'],  inplace=True)
    print(f"PW_LEVEL corr: {X_train['PW_LEVEL'].corr(y_train):.4f}")

# Attorney features
if 'AGENT_ATTORNEY_LAST_NAME' in X_train.columns:
    X_train['HAS_ATTORNEY'] = X_train['AGENT_ATTORNEY_LAST_NAME'].notna().astype(int)
    X_test['HAS_ATTORNEY']  = X_test['AGENT_ATTORNEY_LAST_NAME'].notna().astype(int)
    X_train['ATTY_CERT_RATE'], X_test['ATTY_CERT_RATE'] = smoothed_rate(
        X_train, y_train, X_test, 'AGENT_ATTORNEY_LAST_NAME')
    X_train.drop(columns=['AGENT_ATTORNEY_LAST_NAME'], inplace=True)
    X_test.drop(columns=['AGENT_ATTORNEY_LAST_NAME'],  inplace=True)
    print(f"HAS_ATTORNEY   corr: {X_train['HAS_ATTORNEY'].corr(y_train):.4f}")
    print(f"ATTY_CERT_RATE corr: {X_train['ATTY_CERT_RATE'].corr(y_train):.4f}")

# STATE_MISMATCH: worksite != employer state — third-party placement signal
if 'WORKSITE_STATE' in X_train.columns:
    X_train['STATE_MISMATCH'] = (X_train['EMPLOYER_STATE'] != X_train['WORKSITE_STATE']).astype(int)
    X_test['STATE_MISMATCH']  = (X_test['EMPLOYER_STATE']  != X_test['WORKSITE_STATE']).astype(int)
    X_train.drop(columns=['WORKSITE_STATE'], inplace=True)
    X_test.drop(columns=['WORKSITE_STATE'],  inplace=True)
    print(f"STATE_MISMATCH corr: {X_train['STATE_MISMATCH'].corr(y_train):.4f}")

# WAGE_VS_SOC_MEDIAN: wage vs occupation median (computed while SOC_TITLE still present)
soc_median_map = X_train.groupby('SOC_TITLE')['ANNUAL_WAGE_CAPPED'].median().to_dict()
train_soc_med = X_train['SOC_TITLE'].map(soc_median_map).fillna(X_train['ANNUAL_WAGE_CAPPED'].median())
test_soc_med  = X_test['SOC_TITLE'].map(soc_median_map).fillna(X_train['ANNUAL_WAGE_CAPPED'].median())
X_train['WAGE_VS_SOC_MEDIAN'] = X_train['ANNUAL_WAGE_CAPPED'] - train_soc_med
X_test['WAGE_VS_SOC_MEDIAN']  = X_test['ANNUAL_WAGE_CAPPED']  - test_soc_med
print(f"WAGE_VS_SOC_MEDIAN corr: {X_train['WAGE_VS_SOC_MEDIAN'].corr(y_train):.4f}")

# Safety: drop any unhandled string columns that would crash StandardScaler
extra_str = X_train.select_dtypes(include='object').columns.difference(
    ['EMPLOYER_NAME', 'SOC_TITLE', 'EMPLOYER_STATE']).tolist()
if extra_str:
    print(f'Dropping unhandled string columns: {extra_str}')
    X_train.drop(columns=extra_str, inplace=True)
    X_test.drop(columns=extra_str,  inplace=True)

print(f'\nTotal features before encoding: {X_train.shape[1]}')
```

### Target Encoding

Employer name, SOC title, and employer state each have thousands of unique values. We replace each category with its mean certification rate from the training data. Smoothing with k=50 prevents the model from treating a single denial from a small employer as a strong signal by pulling rare categories toward the global mean.

```python
X_train, X_test = X_train.copy(), X_test.copy()
gm, enc_maps = float(y_train.mean()), {}

def smooth_encode(train_df, test_df, col, target_vals, smoothing=50):

    tmp = train_df[[col]].copy()
    tmp['_y'] = target_vals.loc[train_df.index].values  # index-aligned
    agg  = tmp.groupby(col)['_y'].agg(['mean', 'count'])
    smap = ((agg['count'] * agg['mean'] + smoothing * gm) / (agg['count'] + smoothing)).to_dict()
    enc_maps[col] = smap
    return (train_df[col].astype(str).map(smap).fillna(gm),
            test_df[col].astype(str).map(smap).fillna(gm))

for col in ['EMPLOYER_NAME', 'SOC_TITLE', 'EMPLOYER_STATE']:
    X_train[col + '_ENC'], X_test[col + '_ENC'] = smooth_encode(X_train, X_test, col, y_train)
    X_train.drop(columns=[col], inplace=True)
    X_test.drop(columns=[col],  inplace=True)

remaining_str = X_train.select_dtypes(include='object').columns.tolist()
if remaining_str:
    raise ValueError(f'String columns still present — will crash StandardScaler: {remaining_str}')

FEATURES_ENC = list(X_train.columns)
print(f'Encoding done. Total features: {len(FEATURES_ENC)}')
print(FEATURES_ENC)
```

### Feature Scaling

We normalize all features using StandardScaler fit on the training set only. SMOTE was originally used to oversample denial cases but was removed after the leakage fix weakened the feature set, making synthetic denial cases add noise rather than signal. We use scale_pos_weight in XGBoost instead to weight denial cases more heavily.

```python
scaler  = StandardScaler()
X_tr_sc = scaler.fit_transform(X_train)
X_te_sc = scaler.transform(X_test)


spw = float(y_train.sum() / (y_train == 0).sum())  # will be ~12.6

X_tr_sm = X_tr_sc   # raw scaled training data, no oversampling
y_tr_sm = y_train    # original labels

print(f'Train: Certified={y_train.sum():,} , Not Certified={(y_train==0).sum():,}')
print(f'scale_pos_weight = {spw:.2f}')
print(f'Test set untouched: {len(X_te_sc):,} rows')
print('SMOTE removed — using scale_pos_weight in XGBoost instead.')
```

### Logistic Regression

Logistic Regression is trained as a linear baseline with class_weight balanced so it does not just predict certified for every case. It ended up being competitive with XGBoost, achieving the highest AUC of 0.804.

```python
lr = LogisticRegression(max_iter=500, class_weight='balanced', random_state=42, n_jobs=-1)
lr.fit(X_tr_sm, y_tr_sm)
lr_prob = lr.predict_proba(X_te_sc)[:, 1]
lr_pred = (lr_prob >= 0.5).astype(int)


print("MODEL 1: LOGISTIC REGRESSION")

print(classification_report(y_test, lr_pred, target_names=['Not Certified', 'Certified'], digits=4))
print(f"AUC-ROC: {roc_auc_score(y_test, lr_prob):.4f}")
```

### Random Forest

Random Forest trains 200 decision trees and combines their predictions. We use balanced_subsample class weighting which rebalances within each tree rather than globally, which improves recall on imbalanced data.

```python
rf = RandomForestClassifier(n_estimators=200, max_depth=15, min_samples_leaf=20,
     class_weight='balanced_subsample', random_state=42, n_jobs=-1)
rf.fit(X_tr_sm, y_tr_sm)
rf_prob = rf.predict_proba(X_te_sc)[:, 1]
rf_pred = (rf_prob >= 0.5).astype(int)

print('MODEL 2: RANDOM FOREST')
print(classification_report(y_test, rf_pred, target_names=['Not Certified', 'Certified'], digits=4))
print(f'AUC-ROC: {roc_auc_score(y_test, rf_prob):.4f}')
```

### XGBoost

XGBoost trains trees sequentially where each tree corrects the errors of the previous one. After fixing the data leakage the AUC dropped from 0.917 to around 0.78, which reflects the true difficulty of the problem. We increased max_depth to 7, reduced min_child_weight to 3, and set scale_pos_weight to 5 to focus more on denial cases.

All predicted probabilities cluster between 0.677 and 0.901 due to the high base certification rate. This means the default threshold of 0.50 catches zero denials, so threshold tuning is necessary.

```python
xgb = XGBClassifier(
    n_estimators=1000,
    max_depth=7,
    learning_rate=0.02,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=3,
    reg_alpha=0.1,
    reg_lambda=1.0,
    scale_pos_weight=5,        # moderate — 12.6 was too aggressive
    eval_metric='auc',
    early_stopping_rounds=50,
    random_state=42,
    n_jobs=-1,
    tree_method='hist'
)

xgb.fit(X_tr_sm, y_tr_sm, eval_set=[(X_te_sc, y_test)], verbose=100)
xgb_prob = xgb.predict_proba(X_te_sc)[:, 1]
xgb_pred = (xgb_prob >= 0.5).astype(int)

print('MODEL 3: XGBOOST (threshold = 0.50)')
print(classification_report(y_test, xgb_pred,
      target_names=['Not Certified', 'Certified'], digits=4))
print(f'AUC-ROC: {roc_auc_score(y_test, xgb_prob):.4f}')
print(f'Best iteration: {xgb.best_iteration}')
```

### Threshold Tuning

Because all XGBoost probabilities fall between 0.677 and 0.901, the standard threshold of 0.50 flags nothing as a denial. We search thresholds from 0.05 to 0.92 in fine steps to find the right operating point in the 0.88 to 0.91 range. We use F-beta with beta=2 which weights recall twice as much as precision. The final threshold of 0.895 catches 55.4% of denials with a precision of 22.3%.

```python
thresholds = np.arange(0.05, 0.92, 0.005)
recall_d = [recall_score(y_test,   (xgb_prob>=t).astype(int), pos_label=0, zero_division=0) for t in thresholds]
prec_d   = [precision_score(y_test,(xgb_prob>=t).astype(int), pos_label=0, zero_division=0) for t in thresholds]
fbeta_d  = [fbeta_score(y_test,    (xgb_prob>=t).astype(int), beta=2, pos_label=0, zero_division=0) for t in thresholds]

fbeta_t = thresholds[np.argmax(fbeta_d)]

RECALL_TARGET = 0.55
valid = [(t, r, p) for t, r, p in zip(thresholds, recall_d, prec_d) if r >= RECALL_TARGET]

best_r = recall_score(y_test,    (xgb_prob >= fbeta_t).astype(int), pos_label=0)
best_p = precision_score(y_test, (xgb_prob >= fbeta_t).astype(int), pos_label=0, zero_division=0)

if valid:
    best_t, best_r, best_p = sorted(valid, key=lambda x: x[0], reverse=True)[0]
    print(f'Selected threshold: {best_t:.3f} | Recall: {best_r:.4f} | Precision: {best_p:.4f}')
else:
    best_t = fbeta_t
    print(f'Recall target not achievable. Using F-beta threshold: {best_t:.3f}')

xgb_tuned = (xgb_prob >= best_t).astype(int)
xgb_fbeta = (xgb_prob >= fbeta_t).astype(int)

if xgb_tuned.sum() < len(y_test) * 0.10:
    print('WARNING: Model collapsed. Falling back to F-beta threshold.')
    best_t    = fbeta_t
    xgb_tuned = xgb_fbeta.copy()

print(classification_report(y_test, xgb_tuned, target_names=['Not Certified', 'Certified'], digits=4))
print(f'AUC-ROC: {roc_auc_score(y_test, xgb_prob):.4f}')
```

### Model Comparison

All three models are evaluated on the same test set. Denial recall is the primary metric since missing a denial is more costly than a false alarm. AUC-ROC measures overall discrimination ability independent of threshold.

```python
rows = []
for name, pred, prob in [
    ('Logistic Regression',            lr_pred,   lr_prob),
    ('Random Forest',                  rf_pred,   rf_prob),
    ('XGBoost (t=0.50)',               xgb_pred,  xgb_prob),
    (f'XGBoost F-beta ({fbeta_t:.2f})',xgb_fbeta, xgb_prob),
    (f'XGBoost RC ({best_t:.2f})',     xgb_tuned, xgb_prob),
]:
    rows.append({
        'Model'           : name,
        'Accuracy'        : f'{accuracy_score(y_test, pred)*100:.2f}%',
        'AUC-ROC'         : f'{roc_auc_score(y_test, prob):.4f}',
        'F1-Certified'    : f'{f1_score(y_test, pred, pos_label=1):.4f}',
        'F1-Denial'       : f'{f1_score(y_test, pred, pos_label=0):.4f}',
        'Recall-Denial'   : f'{recall_score(y_test, pred, pos_label=0):.4f}',
        'Precision-Denial': f'{precision_score(y_test, pred, pos_label=0, zero_division=0):.4f}',
    })

results_df = pd.DataFrame(rows)

print('  MODEL COMPARISON SUMMARY')
print(results_df.to_string(index=False))
print('RC = Recall-Constrained threshold | F-beta = F-beta(2) threshold')
```

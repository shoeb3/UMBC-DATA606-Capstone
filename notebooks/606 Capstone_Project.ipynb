##Data Cleaning

```python
print(f"Unique VISA_CLASS values:\n{df['VISA_CLASS'].value_counts()}")
df = df[df['VISA_CLASS'] == 'H-1B'].copy()
print(f"\n After H-1B filter: {len(df):,} rows")
```

```python
print("Original CASE_STATUS distribution:")
print(df['CASE_STATUS'].value_counts())

df['TARGET'] = (df['CASE_STATUS'] == 'Certified').astype(int)

print(f"\n Binary TARGET created:")
print(df['TARGET'].value_counts().rename({1: 'Certified (1)', 0: 'Not Certified (0)'}))
```

```python
# Identifying columns above the 50% threshold
high_null_cols = report[report['Percentage'] > 50].index.tolist()

# Defining specific columns to remove for model clarity
id_cols = ['CASE_NUMBER', 'VISA_CLASS', 'PREPARER_LAST_NAME', 'PREPARER_FIRST_NAME', 'PREPARER_MIDDLE_INITIAL']

# Combining and dropping columns
cols_to_drop = set(high_null_cols + id_cols)
df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)

print(f"Dropped {len(cols_to_drop)} columns.")
print(f"New DataFrame shape: {df.shape}")
```

```python
# Convert Date Columns
date_cols = ['RECEIVED_DATE', 'DECISION_DATE', 'BEGIN_DATE', 'END_DATE']

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Convert Wage Columns
wage_cols = ['WAGE_RATE_OF_PAY_FROM', 'WAGE_RATE_OF_PAY_TO', 'PREVAILING_WAGE']

for col in wage_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('$', '', regex=False)
        df[col] = df[col].astype(str).str.replace(',', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')


all_cols = date_cols + wage_cols
existing_cols = [c for c in all_cols if c in df.columns]
print("Dates and Wages are now numbers.")
print(df[existing_cols].dtypes)
```

```python
unit_map = {
    'Year'      : 1,
    'Month'     : 12,
    'Bi-Weekly' : 26,
    'Week'      : 52,
    'Hour'      : 2080,
}

if 'WAGE_UNIT_OF_PAY' in df.columns:
    df['WAGE_MULTIPLIER'] = df['WAGE_UNIT_OF_PAY'].map(unit_map).fillna(1)
    df['ANNUAL_WAGE'] = df['WAGE_RATE_OF_PAY_FROM'] * df['WAGE_MULTIPLIER']
    df['ANNUAL_PREVAILING_WAGE'] = df['PREVAILING_WAGE']
else:
    df['ANNUAL_WAGE'] = df['WAGE_RATE_OF_PAY_FROM']
    df['ANNUAL_PREVAILING_WAGE'] = df['PREVAILING_WAGE']

print(" Annual wages computed.")
print(df[['ANNUAL_WAGE', 'ANNUAL_PREVAILING_WAGE']].describe())
```

```python
# Calculate the Wage Ratio (Offered Wage / Minimum Required Wage)

df['WAGE_RATIO'] = df['ANNUAL_WAGE'] / df['ANNUAL_PREVAILING_WAGE']

# Calculate how many days it took for the government to decide We just subtract the dates and use '.dt.days' to turn the answer into a number
df['PROCESSING_DAYS'] = (df['DECISION_DATE'] - df['RECEIVED_DATE']).dt.days

# Create a 1 (Yes) or 0 (No) flag for Full-Time position If the column equals 'Y', it becomes 1. Otherwise, it is 0.
df['IS_FULL_TIME'] = (df['FULL_TIME_POSITION'] == 'Y').astype(int)

# Create a 1 (Yes) or 0 (No) flag if the wage is higher than the minimum If Offered Wage is bigger than Prevailing Wage, it becomes 1.
df['WAGE_ABOVE_PREVAILING'] = (df['ANNUAL_WAGE'] > df['ANNUAL_PREVAILING_WAGE']).astype(int)

# Show a summary of our new columns to check the work
print("New features created for analysis!")
print(df[['WAGE_RATIO', 'PROCESSING_DAYS', 'IS_FULL_TIME', 'WAGE_ABOVE_PREVAILING']].describe())
```

```python
# Find the "Cut-off" points for wages We find the bottom 1% (Q1) and the top 1% (Q99) to identify extreme outliers
low_limit = df['ANNUAL_WAGE'].quantile(0.01)
high_limit = df['ANNUAL_WAGE'].quantile(0.99)

print(f"Wages below ${low_limit:,.0f} and above ${high_limit:,.0f} are considered outliers.")

# Clip the wages to these limits This keeps the data but brings extreme values back to the 1% or 99% line
df['ANNUAL_WAGE_CAPPED'] = df['ANNUAL_WAGE'].clip(lower=low_limit, upper=high_limit)

# Print the final dataset size
print(f"\nFinal cleaned dataset has {len(df):,} rows and {df.shape[1]} columns.")

# Count how many cases were successful (Certified) We assume 'TARGET' is 1 for Certified and 0 for Denied
total_cases = len(df)
certified = df['TARGET'].sum()
denied = total_cases - certified

# Calculate percentages simply
perc_certified = (certified / total_cases) * 100
perc_denied = (denied / total_cases) * 100

print(f"Certified: {certified:,} ({perc_certified:.1f}%)")
print(f"Denied: {denied:,} ({perc_denied:.1f}%)")
```

##Target Variable Distribution
Understanding class balance is critical before modeling. Severe imbalance may require resampling strategies such as SMOTE or class-weight adjustments.

```python
target_names = df['TARGET'].replace({1: 'Certified', 0: 'Denied'})

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.countplot(x=target_names, ax=axes[0], palette=['#a6cee3', '#fdbf6f'])
axes[0].set_title('Number of Applications')
axes[0].set_xlabel('Outcome')

counts = target_names.value_counts()
axes[1].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=['#a6cee3', '#fdbf6f'])
axes[1].set_title('Percentage Share')

plt.suptitle('H-1B Visa Outcomes (FY2025)', fontsize=16)
plt.tight_layout()
plt.show()

print("Interpretation: Most cases are Certified (92.6%). Only a small portion are Denied (7.4%).")
```

##Case Status Breakdown (All Categories)
Before binarizing the target, it is informative to see the raw distribution of all CASE_STATUS categories to understand how many records are 'Certified - Withdrawn' vs pure 'Denied'.

```python
all_statuses = []
for d in [df1, df2, df3, df4]:
    h1b_only = d[d['VISA_CLASS'] == 'H-1B']
    all_statuses.append(h1b_only['CASE_STATUS'])

combined_status = pd.concat(all_statuses)
status_counts = combined_status.value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=status_counts.index, y=status_counts.values, palette='Paired')

plt.title('H-1B Applications by Status (FY2025)', fontsize=14, fontweight='bold')
plt.ylabel('Number of Applications')
plt.xticks(rotation=20)

for i, count in enumerate(status_counts.values):
    plt.text(i, count + 1000, f'{count:,}', ha='center')

plt.tight_layout()
plt.show()

print("Interpretation: Most cases are Certified. 'Withdrawn' cases are petitions cancelled by the company.")
```

##Annual Wage Distribution
Salary offered is one of the most important features in the new beneficiary-centric H-1B model. We examine the distribution and compare certified vs. not-certified petitions.

```python
wage_df = df[df['ANNUAL_WAGE_CAPPED'].between(30000, 300000)].copy()

fig, axes = plt.subplots(2, 1, figsize=(12, 10))

sns.histplot(wage_df['ANNUAL_WAGE_CAPPED'], bins=50, ax=axes[0], color='#a6cee3')
axes[0].axvline(wage_df['ANNUAL_WAGE_CAPPED'].median(), color='red', linestyle='--')
axes[0].axvline(wage_df['ANNUAL_WAGE_CAPPED'].mean(), color='orange', linestyle='--')
axes[0].set_title('Distribution of Annual Offered Wages')

sns.kdeplot(data=wage_df, x='ANNUAL_WAGE_CAPPED', hue='TARGET', ax=axes[1], fill=True, palette=['#fdbf6f', '#a6cee3'])
axes[1].set_title('Wage Density: Certified vs. Denied')

plt.tight_layout()
plt.show()

print(wage_df.groupby('TARGET')['ANNUAL_WAGE_CAPPED'].describe())
```

##Wage Ratio Analysis
The Wage Ratio (offered wage / DOL prevailing wage) is central to the 2026 regulatory changes. Under the new beneficiary-centric model, petitions with higher wage ratios receive selection priority. A ratio > 1.0 means the employer is paying above the DOL-mandated prevailing wage.

```python
ratio_df = df[df['WAGE_RATIO'].between(0.5, 3.0)].copy()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(ratio_df['WAGE_RATIO'], bins=50, ax=axes[0], color='#a6cee3')
axes[0].axvline(1.0, color='red', linestyle='--')
axes[0].set_title('Wage Ratio Distribution')

sns.boxplot(data=ratio_df, x='TARGET', y='WAGE_RATIO', ax=axes[1], palette=['#fdbf6f', '#a6cee3'])
axes[1].axhline(1.0, color='red', linestyle='--')
axes[1].set_title('Wage Ratio: Certified (1) vs Denied (0)')

plt.suptitle('H-1B Wage Ratio Analysis (FY2025)', fontsize=16)
plt.tight_layout()
plt.show()

print(ratio_df.groupby('TARGET')['WAGE_RATIO'].describe())
```

## Geographic Analysis: Applications by State
H-1B petitions are heavily concentrated in technology hubs. We examine which states receive the most applications and have the highest certification rates.

##Quarterly Trends (FY2025)
LCA filing volume and wage levels may vary across quarters due to H-1B lottery deadlines (typically April), end-of-fiscal-year rushes, and economic conditions.

```python
date_col = 'RECEIVED_DATE'

if date_col in df.columns:
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    df['QUARTER'] = df[date_col].dt.quarter


fig, axes = plt.subplots(1, 3, figsize=(16, 5))


q_vol = df['QUARTER'].value_counts().sort_index()
sns.barplot(x=q_vol.index.astype(int), y=q_vol.values, ax=axes[0], color='#a6cee3')
axes[0].set_title('Applications per Quarter')
axes[0].set_xlabel('Quarter')


q_cert = df.groupby('QUARTER')['TARGET'].mean() * 100
sns.barplot(x=q_cert.index.astype(int), y=q_cert.values, ax=axes[1], color='#b2df8a')
axes[1].set_title('Success Rate (%)')
axes[1].set_xlabel('Quarter')
axes[1].set_ylim(0, 100)


q_wage = df.groupby('QUARTER')['ANNUAL_WAGE_CAPPED'].median()
sns.barplot(x=q_wage.index.astype(int), y=q_wage.values, ax=axes[2], color='#fdbf6f')
axes[2].set_title('Median Wage per Quarter')
axes[2].set_xlabel('Quarter')

plt.suptitle('H-1B FY2025 Quarterly Analysis', fontsize=16, y=1.05)
plt.tight_layout()


plt.show()
```
